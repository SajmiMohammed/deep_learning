```python
# img ===>grey_scale
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# Reading cascade files
face_cascade=cv2.CascadeClassifier('/home/sajmi/Desktop/BD_MAY/face_detection/data/haarcascade_frontalface_default.xml')
```

```python
# detecting faces only and readed in faces
faces=face_cascade.detectMultiScale(img_grey)
```

```python
len(faces)   # no of faces    
# this particular cascade file has lower accuracy,trained with lower amount of data,
```

```python
faces   # x,y,width,height
```

```python
#Drawing rectagle for each faces
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),color=(0,255,0),thickness=3)
```

```python
cv2.imshow('face detection',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
