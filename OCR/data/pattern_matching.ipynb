```python
# pattern matching
import cv2
import pytesseract
```

```python
from pytesseract import Output
```

```python
data=pytesseract.image_to_data(img,output_type=Output.DICT)
```

```python
data.keys()
```

```python
data['text']
```

```python
data['conf']
```

```python
n_boxes=len(data['text'])
n_boxes
```

```python
for i in range(n_boxes):
        if data['conf'][i]>=70:
            x,y,w,h=data['left'][i],data['top'][i],data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(100,255,0),3)
cv2.imshow('text',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import re
```

```python
# We need only date and email only,here date and email pattern dowmloaded from google(stack overflow)

img=cv2.imread('/home/sajmi/Desktop/BD_MAY/OCR/data/invoice.jpeg')
date_pattern='^(0?[1-9]|[12][0-9]|3[01])[\/\-](0?[1-9]|1[012])[\/\-]\d{4}$'
email_pattern='^[a-z0-9](\.?[a-z0-9]){5,}@example\.net$'
for i in range(n_boxes):
    if data['conf'][i]>=70:     
        if re.match(date_pattern,data['text'][i]):   # matching date
            x,y,w,h=data['left'][i],data['top'][i],data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(100,255,0),3)
cv2.imshow('text1',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# matching email

img=cv2.imread('/home/sajmi/Desktop/BD_MAY/OCR/data/invoice.jpeg')

for i in range(n_boxes):

    if data['conf'][i]>=70:
        
        if re.match(email_pattern,data['text'][i]):
            x,y,w,h=data['left'][i],data['top'][i],data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(100,255,0),3)
cv2.imshow('email',img)
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
# date and mail together

img=cv2.imread('/home/sajmi/Desktop/BD_MAY/OCR/data/invoice.jpeg')

for i in range(n_boxes):

    if data['conf'][i]>=70:

        if re.match(email_pattern,data['text'][i]) or re.match(date_pattern,data['text'][i]):
            
            x,y,w,h=data['left'][i],data['top'][i],data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(100,255,0),3)
cv2.imshow('emaildate',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
